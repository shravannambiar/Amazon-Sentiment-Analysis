## splitting the data into training and testing  

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3)
```

## performing cross validation to find the optimum value of k

```python
neighbors=list(range(1,20,2))
cv_score=[]
for k in neighbors:
    knn=KNeighborsClassifier(n_neighbors=k,n_jobs=4)
    scores=cross_val_score(knn,X_train,Y_train,cv=5,scoring="accuracy")
    cv_score.append(scores.mean())
    print(cv_score)  
```

```python
MSE=[1-x for x in cv_score]
```

```python
MSE
```

## plotting neighbors against error to visualize 

```python
plt.plot(neighbors,MSE)
plt.xlabel("neighbors")
plt.ylabel("error")
```

## finding optimal k which is the value of neighbor for which we found the least error

```python
optimal_k=neighbors[MSE.index(min(MSE))]
```

```python
optimal_k
```

## building the classifier with this optimal k value

```python
classifier=KNeighborsClassifier(n_neighbors=optimal_k,n_jobs=4)
```

## tarining the classifier

```python
classifier.fit(X_train,Y_train)
```

## testing the classifier

```python
y_pred=classifier.predict(X_test)
```

```python
acc=accuracy_score(Y_test,y_pred)
```
