## final.sqlite contains preprocessed data and is shuffled with random state =123

```python
con=sqlite3.connect("saved_data//final.sqlite")
final=pd.read_sql_query("select * from Reviews",con)
con.close
```

```python
tfidf=TfidfVectorizer()
tfidf_vector=tfidf.fit_transform(final["CleanedText"].values)
```

# taking subset of first 2000 points

```python
df1=final["Score"][200000:202000]
```

```python
df1
```

```python
df2=pd.DataFrame(tfidf_vector[200000:202000,:].toarray())
```

```python
df2
```

```python
df=pd.concat([df1,df2],axis=1)
```

```python
df
```

```python
Y=df["Score"]
```

```python
X=df.iloc[:,1:].values
```

```python
s=StandardScaler()
X=s.fit_transform(X)
```

```python
X.shape
```

# splitting into test and train parts

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size = 0.3, random_state = 123)
```

```python
?KNeighborsClassifier
```

```python
?cross_val_score
```

### finding the optimal value of k

```python
cv_scores=[]
neighbors=list(range(1,20,2))
for k in neighbors:
    knn=KNeighborsClassifier(n_neighbors=k,n_jobs=4)
    score=cross_val_score(knn,X_train,Y_train,cv=5,scoring="accuracy")
    print(score)
    cv_scores.append(score.mean())
    
```

```python
MSE=[1-x for x in cv_scores]
```

```python
optimal_knn=neighbors[MSE.index(min(MSE))]
optimal_knn
```

### training the model

```python
classifier = KNeighborsClassifier(n_neighbors = optimal_knn,n_jobs=5)
classifier.fit(X_train,Y_train)
```

### testing the model

```python
y_pred = classifier.predict(X_test)
```

```python
?accuracy_score
```

### finding accuracy

```python
acc = accuracy_score(Y_test, y_pred, normalize=True) * float(100)
acc
```
